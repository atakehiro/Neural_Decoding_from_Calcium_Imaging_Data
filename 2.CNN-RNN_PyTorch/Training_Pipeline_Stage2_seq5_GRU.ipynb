# Stage2 Training

```python
from fastai_stage2_training_efficientnet_rnn import stage2_train
import itertools
```

```python
RNN_Type = 'GRU'
CNN_fix = 'nonFix' #'Fix' or 'nonFix'
SEQ_SIZE = 5 # input frame number [-n,n+1]
```

```python
l = [1, 2, 3, 4, 5]
for v in itertools.permutations(l, 2):
    TRAIN = [i for i in l if i not in v]
    VALID = [v[0]]
    TEST = [v[1]]
    stage2_train(TRAIN, VALID, TEST, RNN_Type, CNN_fix, SEQ_SIZE)
```
